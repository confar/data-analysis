# how to filter rows of a table by column value

```python
import pandas as pd
pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.shape
```

```python
booleans = []
for length in movies.duration:
    if length > 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
is_long = movies.duration >= 200
is_long.head()
```

```python
is_long = pd.Series(booleans)
```

```python
is_long.head()
```

```python
movies.loc[movies.duration >= 200, 'genre']
```

# Titanic dataframe Kaggle. How do I use pandas with scikit-learn

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head()
```

```python
feature_cols = ['Pclass', 'Parch']
```

```python
x = train.loc[:, feature_cols]
```

```python
x.shape
```

```python
y = train.Survived
```

```python
y.shape
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x,y)
```

```python
test = pd.read_csv('http://bit.ly/kaggletest')
```

```python
test.head()
```

```python
x_new = test.loc[:, feature_cols]
```

```python
x_new.shape
```

```python
new_pred_class = logreg.predict(x_new)
```

```python
new_pred_class
```

```python
pd.DataFrame({'PassengerId': test.PassengerId, 'Survived':new_pred_class}).set_index('PassengerId')
```
